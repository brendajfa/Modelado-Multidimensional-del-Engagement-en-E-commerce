```python
import pandas as pd
import os

from src.data_loader import general_cleaning
from src import config

config.WORKPATH = "drive/MyDrive/Master Data Science/TFM/GITHUB"
```

```python
workpath = config.WORKPATH
```

```python
df_raw = pd.read_parquet(os.path.join(workpath, "raw", "2019-Nov.parquet"))
df_cleaned = general_cleaning(df_raw)
df_cleaned = pd.read_parquet(os.path.join(workpath, "raw", "2019-Nov-cleaned.parquet"))
df = df_cleaned.dropna(subset=["category_code"])
```

```python
for category_code in config.CATEGORIES_CODES:
  df_ = df[df["category_code"] == category_code]
  df_ = df_.drop_duplicates(["event_type", "user_session", "product_id", "price"])
  df_.to_parquet(
    os.path.join(workpath, "categories", "training_201911", f"{category_code}.parquet")
  )
```
