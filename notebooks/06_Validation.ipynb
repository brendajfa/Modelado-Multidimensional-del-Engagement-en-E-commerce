```python
from src.evaluation.absence_time import compute_absence_time_summary

summary_absence_time_nov = compute_absence_time_summary(
    config.WORKPATH,
    source_folder="training_201911",
    category_codes=config.CATEGORY_CODES,
    output_path=os.path.join(config.RESULTSPATH, "validation_analysis", "training_AbsenseTime.parquet"),
    show_progress=False
)
summary_absence_time_dec = compute_absence_time_summary(
    config.WORKPATH,
    source_folder="validation_201912",
    category_codes=config.CATEGORY_CODES,
    output_path=os.path.join(config.RESULTSPATH, "validation_analysis", "evaluation_AbsenseTime.parquet"),
    show_progress=False
)
```

```python
summary_absence_time_nov = pd.read_parquet(os.path.join(config.RESULTSPATH, "validation_analysis", "training_AbsenseTime.parquet"))
summary_absence_time_dec = pd.read_parquet(os.path.join(config.RESULTSPATH, "validation_analysis", "evaluation_AbsenseTime.parquet"))
cluster_user_based_nov = pd.read_parquet(os.path.join(config.RESULTSPATH, "clusters", "training_user_clusters.parquet"))
cluster_user_based_dec = pd.read_parquet(os.path.join(config.RESULTSPATH, "clusters", "validation_user_clusters.parquet"))
```

```python
cols = ["category_code", "cluster"]
summary_absence_time_nov = summary_absence_time_nov.merge(cluster_user_based_nov[cols], on="category_code", how="left").rename(columns={"cluster": "user_based_nov"})
summary_absence_time_nov = summary_absence_time_nov.merge(cluster_user_based_dec[cols], on="category_code", how="left").rename(columns={"cluster": "user_based_dic"})

comparison = summary_absence_time_nov.merge(summary_absence_time_dec, on="category_code", how="left", suffixes=('_nov','_dic'))
comparison['absence_drop'] = comparison['absence_time_mean_dic'] - comparison['absence_time_mean_nov']

df_clean = comparison.drop_duplicates(subset=['category_code', 'absence_time_mean_dic'], keep='last')
df_clean = df_clean.reset_index(drop=True)

df_clean['cluster'] = df_clean['user_based_dic']
```

```python
cluster_summary = df_clean.groupby('cluster').agg({
    'absence_time_mean_nov': 'mean',
    'absence_time_mean_dic': 'mean',
    'absence_drop': 'mean'
})
cluster_summary
```

```python
urgent_categories = df_clean[df_clean['absence_drop'] < 0].sort_values('absence_drop')
urgent_categories[['category_code','absence_drop','cluster']]
```
